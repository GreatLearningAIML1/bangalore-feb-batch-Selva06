```python
#Question2: Answer

old_mean=df['Old Scheme (in thousands)'].mean()
new_mean=df['New Scheme (in thousands)'].mean()
# old scheme mean
print (old_mean)
# New scheme mean
new_mean

#The data is dependent data. Since the different test is performed on the same set of population
# Old and New Scheme has two different means, which is indicatiing the variance in the distribution
#two groups doesnt have the same mean at the 5% significance level
#The null hypothesis is U1=u2
#Alternative hypothesis u1!=u2

#it shows that new scheme is significantly raised outputs
```

```python
group1 = df.loc[:,'Old Scheme (in thousands)']
print (group1)
group2 = df.loc[:,'New Scheme (in thousands)']
group2
```

```python
t_statistic, p_value = ttest_ind(group1, group2)
print(t_statistic, p_value)
```

```python
print ("two-sample t-test p-value=", p_value)
#Question 3:
 #P value= 0.49, which is 49% of the group.
# P value says that there could be 49% of rejection of Null hypothesis.
```

```python
u, p_value = mannwhitneyu(group1, group2)
print ("two-sample wilcoxon-test p-value=", p_value)
```
